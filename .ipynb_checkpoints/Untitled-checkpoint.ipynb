### Step 1 : Use the request libary to download web pages

```python
import requests
```

```python
# get the web page
topics_url = "https://github.com/topics"
```

```python
#download the page
response = requests.get(topics_url)
```

```python
# if the response download was succesfull
response.status_code
```

```python
# get the content on the downlaoded page
page_content = response.text
page_content
```

### Use Beautiful soup to parse and extract information

```python
from bs4 import BeautifulSoup as bs
```

```python
# parse doc
doc = bs(page_content, "html.parser")
```

```python
doc
```

```python
# find specific tags to track collects
p_tags = doc.find_all("p")
```

```python
p_tags
```
